```python
graph = {'A': set(['B', 'E']), 'B': set(['A', 'C', 'F']), 'C': set(['B', 'D', 'G']),
         'D': set(['C', 'H']), 'E': set(['A', 'F', 'I']), 'F': set(['B', 'E', 'I', 'G']),
         'G': set(['C', 'F', 'H']), 'H': set(['D', 'G', 'L']), 'I': set(['E', 'F', 'M']),
         'J': set([]), 'K': set(['L']), 'L': set(['H', 'K', 'P']), 'M': set(['I', 'Q']),
         'N': set(['R']), 'O': set(['P', 'S']), 'P': set(['L', 'T']), 'Q': set(['M', 'U', 'R']),
         'R': set(['N', 'Q', 'V', 'S']), 'S': set(['O', 'R', 'W', 'T']), 'T': set(['P', 'S', 'X']),
         'U': set(['Q', 'V']), 'V': set(['R', 'U', 'W']), 'W': set(['S', 'V', 'X']), 'X': set(['T', 'W'])
         }

# Search that will search all the paths
def dfs_paths(graph, start, goal):
    totalPaths = []
    stack = [(start, [start])]

    while stack:
        (vertex, path) = stack.pop()
        for next in graph[vertex] - set(path):
            if next == goal:
                #print(path + [next])
                totalPaths.append(path + [next])
                #yield path + [next]
            else:
                stack.append((next, path + [next]))

    return totalPaths

# Create a variable to hold the list of paths generated by dfs_paths
# Pass in 3 parameters to dfs_paths
# Graph to traverse, initial state and goal state.
totalPaths = dfs_paths(graph, 'A', 'K')
print('Every Path: ')
print(totalPaths)
print('Total number of paths: ' + str(len(totalPaths)))
if (len(totalPaths)==0):
  print('Goal cannot be reached.')
else:
  shortestPath = min(totalPaths, key=len)
  print('Shortest Path: ' + "-".join(shortestPath))
  print('Shortest Path Length: ' + str(len(shortestPath)))
  print('Shortest Path Cost: ' + str(len(shortestPath)-1))
```
